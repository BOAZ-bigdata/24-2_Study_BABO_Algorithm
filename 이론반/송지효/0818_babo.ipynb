```python
# 1260번

# BFS 는 스택사용 : 처음 선택된 거 넣기 -> 빼서 그거랑 관련된 거 넣기(오름차순)
# DFS 는 처음 선택 -> 거기에서 최적 값 선택 -> 그거랑 관련된 최적 값 선택

# DFS에서는 그냥 받은 거 그대로 딕셔너리로 오름차순으로 저장, 키에 따른 값을 찾으면 됨.
# BFS에서는 받은거 그대로 딕셔너리로 오름차순으로 저장, 해당하는 거 다 쓸때까지 다음으로 절대 못 넘어감.
# 선택된 관련 노드 다 찾아서 삭제할 것 -> 단, dfs와 bfs가 동기화가 되지 않도록 전역변수로 사용하지 않는 것이 중요

from collections import deque

# 입력 받기
N, M, V = map(int, input().split())
graph = [[] for _ in range(N + 1)]

# 그래프 입력 받기
for _ in range(M):
    head, tail = map(int, input().split())
    graph[head].append(tail)
    graph[tail].append(head)

# 그래프의 각 노드에 대해 인접 노드를 오름차순으로 정렬
for i in range(1, N + 1):
    graph[i].sort()

# DFS 함수 정의 (재귀)
def dfs(v, visited):
    visited[v] = True
    print(v, end=' ')

    for neighbor in graph[v]:
        if not visited[neighbor]:
            dfs(neighbor, visited)

# BFS 함수 정의
def bfs(v):
    queue = deque([v])
    visited = [False] * (N + 1)
    visited[v] = True

    while queue:
        node = queue.popleft()
        print(node, end=' ')

        for neighbor in graph[node]:
            if not visited[neighbor]:
                queue.append(neighbor)
                visited[neighbor] = True

# 방문 리스트 초기화 및 DFS 실행
visited_dfs = [False] * (N + 1)
dfs(V, visited_dfs)
print()

# BFS 실행
bfs(V)
```

```python
#11724번
# 재귀함수 이용해서 끝까지 연결되어 있는 지 확인

def dfs(v, visited, graph):
    visited[v] = True
    for neighbor in graph[v]:
        if not visited[neighbor]:
            dfs(neighbor, visited, graph)

# 입력 받기
N, M = map(int, input().split())
graph = [[] for _ in range(N + 1)]

# 그래프 입력 받기
for _ in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

# 방문 리스트 초기화
visited = [False] * (N + 1)

# 연결 요소 카운트 초기화
connected_components = 0

# 모든 정점에 대해 탐색
for i in range(1, N + 1):
    if not visited[i]:
        # 방문하지 않은 노드에서 DFS 시작
        dfs(i, visited, graph)
        connected_components += 1

# 결과 출력
print(connected_components)

```

```python
# 1743번

# 입력 받기
n, m, k = map(int, input().split())

# 그래프 초기화 (n x m)
graph = [[0] * m for _ in range(n)]

# 음식물 위치를 그래프에 표시
for _ in range(k):
    r, c = map(int, input().split())
    graph[r-1][c-1] = 1  # 좌표는 0-indexed로 변환

# 방향 벡터 (상, 하, 좌, 우)
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(x, y):
    stack = [(x, y)]
    size = 0

    while stack:
        x, y = stack.pop()

        # 이미 방문한 노드는 무시
        if graph[x][y] == 0:
            continue

        # 방문 처리 및 크기 증가
        graph[x][y] = 0
        size += 1

        # 인접한 노드를 탐색
        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]
            if 0 <= nx < n and 0 <= ny < m and graph[nx][ny] == 1:
                stack.append((nx, ny))

    return size

# 최대 크기 계산
max_size = 0

for i in range(n):
    for j in range(m):
        if graph[i][j] == 1:
            max_size = max(max_size, dfs(i, j))

# 결과 출력
print(max_size)
```
