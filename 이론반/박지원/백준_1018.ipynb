### 1번
[백준 1018](https://www.acmicpc.net/problem/1018)

문제 해결 아이디어 및 고려할점

1. 8X8 체스판 가능한 경우 2가지 A, B로 만들고 for문으로 한칸씩 오른쪽, 아래쪽으로 계산하면서 최소의 경우 return 

2. brute force 알고리즘을 사용하면 되지만, 시간복잡도를 고려해야하는 문제

- 시간 복잡도 계산하면 N x M x 64 -> O(NM)

```python
# 입력받아오기
N, M = map(int, input().split())
board = [list(input()) for _ in range(N)]
res = []

# 체스판 가능한 경우 2가지 정의 
board_A = [['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W']]

board_B = [['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'],
           ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W'],
           ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B']]

for i in range(N-7):
    for j in range(M-7):
        cnt_A = 0
        cnt_B = 0
        for a in range (8):
            for b in range (8):
                if board[a+i][b+j] != board_A[a][b]: cnt_A += 1 # 해당 위치와 다르면 +1
                if board[a+i][b+j] != board_B[a][b]: cnt_B += 1
        
        res.append(cnt_A)
        res.append(cnt_B)
        
                
print(min(res)) # 최솟값 출력     
```
