1. DFS & BFS

```python
num_vertices, num_edges, start_vertex = map(int, input().split())


graph = [[0] * (num_vertices + 1) for _ in range(num_vertices + 1)]


for _ in range(num_edges):
    vertex_a, vertex_b = map(int, input().split())
    graph[vertex_a][vertex_b] = graph[vertex_b][vertex_a] = 1


dfs_visited = [0] * (num_vertices + 1)
bfs_visited = dfs_visited.copy()

def dfs(vertex):
    dfs_visited[vertex] = 1  
    print(vertex, end=' ')
    for adjacent_vertex in range(1, num_vertices + 1):
        if graph[vertex][adjacent_vertex] == 1 and dfs_visited[adjacent_vertex] == 0:
            dfs(adjacent_vertex)


def bfs(vertex):
    queue = [vertex]
    bfs_visited[vertex] = 1  
    while queue:
        current_vertex = queue.pop(0)  
        print(current_vertex, end=' ')
        for adjacent_vertex in range(1, num_vertices + 1):
            if bfs_visited[adjacent_vertex] == 0 and graph[current_vertex][adjacent_vertex] == 1:
                queue.append(adjacent_vertex)
                bfs_visited[adjacent_vertex] = 1 


dfs(start_vertex)
print()
bfs(start_vertex)
```

2. 연결 요소의 개수

```python
import sys


def dfs(vertex):
    global visited
    visited[vertex] = 1  
    stack = [] 
    stack.append(vertex)  
    
    while stack:
        current_vertex = stack.pop()  
        for adjacent_vertex in graph[current_vertex]:  
            if not visited[adjacent_vertex]:  
                visited[adjacent_vertex] = visited[current_vertex] + 1  
                stack.append(adjacent_vertex) 


N, M = map(int, sys.stdin.readline().split())
graph = [[] for _ in range(N + 1)]  
visited = [0] * (N + 1)  


for _ in range(M):
    vertex_u, vertex_v = map(int, sys.stdin.readline().split())
    graph[vertex_u].append(vertex_v)
    graph[vertex_v].append(vertex_u)


component_count = 0
for vertex in range(1, N + 1):
    if not visited[vertex]:  
        dfs(vertex)  
        component_count += 1 
        

print(component_count)
```

3. 음식물 피하기

```python
import sys
from collections import deque


n, m, k = map(int, sys.stdin.readline().split())


graph = [[0] * m for _ in range(n)]
for _ in range(k):
    a, b = map(int, sys.stdin.readline().split())
    graph[a - 1][b - 1] = 1  


visited = [[False] * m for _ in range(n)]


dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]


max_connected_count = 0


def bfs(start_x, start_y):
    global cnt
    queue = deque()
    queue.append((start_x, start_y)) 
    visited[start_x][start_y] = True
    cnt = 1  

    while queue:
        x, y = queue.popleft()  
        for direction in range(4):  
            next_x = x + dx[direction]
            next_y = y + dy[direction]
            if 0 <= next_x < n and 0 <= next_y < m:  
                if graph[next_x][next_y] == 1 and not visited[next_x][next_y]:  
                    visited[next_x][next_y] = True  
                    cnt += 1  
                    queue.append((next_x, next_y))  


for i in range(n):
    for j in range(m):
        if graph[i][j] == 1 and not visited[i][j]:  
            cnt = 0  
            bfs(i, j)  
            max_connected_count = max(max_connected_count, cnt)  

print(max_connected_count)
```

4. 토마토

```python
from collections import deque


m, n = map(int, input().split())


tomato = []
queue = deque()


for i in range(n):
    row = list(map(int, input().split()))
    tomato.append(row)
    
   
    for j in range(m):
        if row[j] == 1:
            queue.append((i, j))


dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]


def bfs():
    while queue:
        x, y = queue.popleft()  
        
  
        for direction in range(4):
            nx, ny = x + dx[direction], y + dy[direction]
            if 0 <= nx < n and 0 <= ny < m:
                if tomato[nx][ny] == 0:
                    tomato[nx][ny] = tomato[x][y] + 1
                    queue.append((nx, ny))  

bfs()

max_days = 0
for i in range(n):
    for j in range(m):
        if tomato[i][j] == 0:
            print(-1)
            exit()
        if tomato[i][j] > max_days:
            max_days = tomato[i][j]

print(max_days - 1)
```
